```python
import numpy as np
import PyTCI
```

```python
model = PyTCI.example.pretrained()
analyzer = PyTCI.Analyzer(model, 100)
```

```python
audio = [(i+1)*np.ones((l, 1)) for i, l in enumerate((1500, 1600, 1800, 1200, 600, 2100, 960, 560, 1300, 1700))]
sr = 100
```

```python
analyzer.set_audio(audio, sr)
```

```python
x1 = analyzer.sequence(0.8, 0)
y1 = analyzer.crossfade(x1)
z1 = model(y1).reshape(10, 80, 16)
t1 = analyzer.desequence(z1, 0.8, 0)

x2 = analyzer.sequence(0.8, 1)
y2 = analyzer.crossfade(x2)
z2 = model(y2).reshape(10, 80, 16)
t2 = analyzer.desequence(z2, 0.8, 1)
```
